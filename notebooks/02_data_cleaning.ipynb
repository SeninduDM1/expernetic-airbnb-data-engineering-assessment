```python
import pandas as pd
```

```python
df = pd.read_csv("data/raw/listings.csv.gz")
```

```python
clean_df = df.copy()
```

```python
clean_df.shape
```

```python
missing_percentage = clean_df.isnull().mean() * 100

columns_to_drop = missing_percentage[
    missing_percentage == 100
].index

clean_df = clean_df.drop(columns=columns_to_drop)
```

```python
clean_df.shape
```

```python
clean_df["price"].head(10)
```

```python
clean_df["price"] = clean_df["price"].str.replace("$", "", regex=False)
```

```python
clean_df["price"].head(10)
```

```python
clean_df["price"].dtype
```

```python
clean_df["price"] = pd.to_numeric(clean_df["price"], errors="coerce")
```

```python
clean_df["price"].dtype
```

```python
clean_df["price"].describe()
```

```python
clean_df["price"].dtype
```

```python
clean_df.to_csv(
    "data/processed/listings_clean.csv",
    index=False
)
```
